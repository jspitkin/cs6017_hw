```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
pm25 = pd.read_csv("2017-PM2.5.csv", header=[0], skiprows=[1,2], index_col=0, parse_dates=True)
#print(pm25.describe())
print(pm25['RP'][:10])
```

```python
rp = pm25['RP']
rp.plot()
```

```python

rp.groupby(lambda x:  x.time()).mean().plot(kind="bar")
```

```python
axes = pm25.boxplot(by = [lambda x: x.time()], column = 'RP', rot=90, showfliers=False)
axes.set_xlabel("Time of Day")
plt.title("PM2.5 concentration")
plt.suptitle("")
plt.show()
```

```python
axes = pm25.boxplot(by = [lambda x: x.month], column = 'RP', rot=90, showfliers=False)
axes.set_xlabel("Month")
plt.title("PM2.5 concentration")
plt.suptitle("")
plt.show()
```

```python
np.corrcoef(np.array(pm25['RP']), np.array(pm25['HW']))
```
